```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("House_prediction.csv")
data_sorted = data.groupby(["city"])[list(data.head(0))]
data_sorted = data_sorted.mean()
print(data_sorted.shape)
data_sorted
```

```python
for i in range(data_sorted.shape[1]):
    plt.plot(data_sorted.index.tolist(),data_sorted.loc[:,list(data_sorted.head(0))[i]])
    plt.ylabel(list(data_sorted.head(0))[i])
    plt.show()
```

```python
corr = data_sorted.corr()
corr.style.background_gradient(cmap='coolwarm')
```
